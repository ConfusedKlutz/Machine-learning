```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
#difflib: matches with the closest value in the dataset, the user might make an error during input and this resolves that
```


# **Data Collection and Pre-processing**

```python
#loading the data from the csv file and to a pandas dataframe
movies_data = pd.read_csv ("movies.csv")
```

```python
#printing the first five rows of the dataframe
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combining all selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']

print(combined_features)
```

```python
#converting the textual data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
#will transform our data into numerical values
print(feature_vectors)
```

```python
#now use the cosine similarity algorithm to get the similarity score
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape) #1 -> here represents that its basically the same element
```

```python
 #getting the movie name as input from the user
movie_name = input('Enter your favourite movie name: ')
```

```python
#create a list with all the movies given in the dataset which will be then be used for comparision with the name provided
list_of_all_movies = movies_data['title'].tolist()
print(list_of_all_movies)
```

```python
#find the closest match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_movies)
print(find_close_match)
```

```python
#we need to extract the closest match, the first(most relevant value)
close_match = find_close_match[0]
print(close_match)
```

```python
#find the index of the movie
index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
#now we'll get a list of similar movies, based on similarity score
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
len (similarity_score)
```

```python
#sort the list
sorted_similar_movies = sorted (similarity_score, key = lambda x:x[1], reverse = True)
print (sorted_similar_movies)
```

```python
#now we need to print the titles corresponding to the indices
print("movies suggested for you: \n")
i=1
for movie in sorted_similar_movies:
  index = movie[0]   #retrieves the indices, not pointing to the first element or something.
  title_from_index = movies_data[movies_data.index==index]['title'].values[0] #select the movie with the index
  if(i<30):
    print(i, '.', title_from_index)
    i+=1



```
